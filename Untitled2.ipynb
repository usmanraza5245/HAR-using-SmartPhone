```python
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

training_activity = train['Activity']
testing_activity = test['Activity']
train = train.iloc[:,:260]
test = test.iloc[:,:260]
#filter data
train = train.drop(train.filter(regex='meanFreq|Jerk'),axis=1)
test = test.drop(test.filter(regex='meanFreq|Jerk'),axis=1)

train = train.filter(regex='mean|std')
test = test.filter(regex='mean|std')
print(train)
print(test)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder().fit(['STANDING','SITTING','LAYING','WALKING','WALKING_UPSTAIRS','WALKING_DOWNSTAIRS'])
training_act_enc = le.transform(training_activity)
testing_act_enc = le.transform(testing_activity)

```

```python
import keras
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils.np_utils import to_categorical
from keras.models import Sequential

# Now let's experiment with a neural network to classify this data and see if we can improve our accuracy even further
# First we need to encode our targets as one-hot label vectors
training_labels = to_categorical(training_act_enc)
testing_labels = to_categorical(testing_act_enc)
```

```python
model = Sequential()
model.add(Dense(16,input_shape=train.shape, activation='relu'))

model.add(Dense(8,activation='relu'))
model.add(Dense(6))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
history = model.fit(train.values, training_labels, verbose = 2, validation_split = .15, shuffle=True)
```
